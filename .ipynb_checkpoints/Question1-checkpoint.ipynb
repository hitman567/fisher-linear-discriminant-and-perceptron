```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import mpld3
mpld3.enable_notebook()
```

```python
data=pd.read_csv("dataset_3.csv")
# data.head()
```

```python
# x=data['X']
# y=data['Y']
# plt.scatter(x,y,c=data["target"])
```

```python
data_groupby=data.groupby("target")
data_groupby.first()

# print(data_groupby)
```

```python
data0=data_groupby.get_group(0)
data1=data_groupby.get_group(1)
```

```python
# data0.head()
```

```python
plt.scatter(data0['X'],data0["Y"],color="r")
plt.scatter(data1['X'],data1["Y"],color="b")
```

```python
mean0=np.mean(data0)
mean1=np.mean(data1)


# print(mean0.shape)
overallmean=np.mean(data.drop(['target'],axis=1))

```

```python
# mean0=mean0.as_matrix()
# mean1=mean1.as_matrix()
# overallmean=overallmean.as_matrix()
```

```python
mean0=np.resize(mean0,(2,1))
mean1=np.resize(mean1,(2,1))
overallmean=np.resize(overallmean,(2,1))
```

```python
# print(mean0.shape)
# print(mean1.shape)
# print(overallmean.shape)
```

```python
n0=data0["X"].count()
n1=data1["X"].count()
# print(n1)
# print(n0)
```

```python
s0=np.matmul((mean0-overallmean)*n0,(mean0-overallmean).T)
# print(s0)
s1=np.matmul((mean1-overallmean)*n1,(mean1-overallmean).T)
# print(s1)
```

```python
sb=s0+s1
sb=np.array(sb)
# print(sb.shape)
# print(sb)
```

```python
sw0=np.zeros(sb.shape)
for i in data0.itertuples():
    mat=np.zeros((2,1))
    mat[0][0]=i.X
    mat[1][0]=i.Y
    sw0=np.add(np.matmul((mat-mean0),(mat-mean0).T),sw0)

    
```

```python
# print(sw0)
```

```python
sw1=np.zeros(sb.shape)
for i in data1.itertuples():
    mat=np.zeros((2,1))
    mat[0][0]=i.X
    mat[1][0]=i.Y
    sw1=np.add(np.matmul((mat-mean1),(mat-mean1).T),sw1)

    
```

```python
# print(sw1)
```

```python
sw=sw0+sw1
# print(sw)
```

```python
swinv=np.linalg.inv(sw)
# print(swinv)
```

```python
direction=np.matmul(swinv,(mean1-mean0))
# print(direction)
# print(direction.shape)
```

```python
proj_zero=[]
# direc=direction[0][0]**2+direction[1][0]**2
# print(direc)


```

```python
for i in data0.itertuples():
    mat=np.zeros((2,1))
    mat[0][0]=i.X
    mat[1][0]=i.Y
    val=mat[0][0]*direction[0][0]+mat[1][0]*direction[1][0]
#     val=val/direc
    proj_zero.append(val)
#     c=np.dot(direction,direction)
```

```python
# print(proj_zero)
```

```python
proj_one=[]
for i in data1.itertuples():
    mat=np.zeros((2,1))
    mat[0][0]=i.X
    mat[1][0]=i.Y
    val=mat[0][0]*direction[0][0]+mat[1][0]*direction[1][0]
#     val=val/direc
    proj_one.append(val)
#     c=np.dot(direction,direction)
```

```python
# print(proj_one)
```

```python

# x=data['X']
# y=data['Y']
# plt.scatter(x,y,c=data["target"])
for i in proj_zero:
    plt.scatter(i,0,color="r")
for i in proj_one:
    plt.scatter(i,0,color="b")
    
```

```python
val=mean0[0][0]*direction[0][0]+mean0[1][0]*direction[1][0]


mean0_proj=val
# print(mean0_proj)
```

```python
val=mean1[0][0]*direction[0][0]+mean1[1][0]*direction[1][0]

mean1_proj=val

mean=(mean1_proj+mean0_proj)/2

# print(mean)
```
