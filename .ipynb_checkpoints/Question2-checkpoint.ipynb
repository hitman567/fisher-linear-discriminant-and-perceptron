```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("dataset_1.csv")
data.head()
```

```python
wt=np.ones((3,1))
y=data["target"]
y=np.array(y)
print(y)
print(y.shape)
y.resize(1000,1)
print(y.shape)
print(wt.shape)
print(wt[2][0])
```

```python
datalist=[]
for i in data.itertuples():
    list=[]
    list.append(0)
    list.append(i.X)
    list.append(i.Y)
    datalist.append(list)
```

```python
datalist
```

```python
plt.figure(figsize=(10,6))
yhat=[]
count=0
# print(wt[2][0])
for i in datalist:
    sum=wt[0][0]*i[0]+wt[1][0]*i[1]+wt[2][0]*i[2]
    if(sum>0):
        yhat.append(1)
    else:
        yhat.append(0)
    wt[0][0]=wt[0][0]+0.1*(y[count]-yhat[count])*i[0]
    wt[1][0]=wt[1][0]+0.1*(y[count]-yhat[count])*i[1]
    wt[2][0]=wt[2][0]+0.1*(y[count]-yhat[count])*i[2]
    count+=1
    slope=-(wt[0][0]/wt[2][0])/(wt[0][0]/wt[1][0])
    y_intercept=-(wt[0][0]/wt[2][0])
#     print(slope)
    x1=i[1]
    y1=slope*x1+y_intercept
#     x2=datalist[-1][1]
#     y2=slope*x2+y_intercept
    plt.ylim(-3,3)
    plt.xlim(-3,3)
    plt.scatter(x1,y1)
```
